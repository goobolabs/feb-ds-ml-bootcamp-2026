```python
# Assignment-06: Spending Pattern Analysis with K-Means

import os
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Optional: suppress MKL warning on Windows
os.environ["OMP_NUM_THREADS"] = "1"

# 1. Load dataset (Excel file in lesson-06/dataset folder)
df = pd.read_csv("spending_l9_dataset.csv")


# 2. Select features
X = df[["Income_$", "SpendingScore"]]

# 3. Handle missing values
X = X.fillna(X.median())

# 4. Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Elbow Method (SSE loop)
print("=== SSE VALUES (Elbow Check) ===")
for k in range(1, 10+1):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    print(f"k={k}, SSE={km.inertia_}")

# 6. Choose K (example: 3 clusters based on elbow)
K = 3
km = KMeans(n_clusters=K, random_state=42)
labels = km.fit_predict(X_scaled)
df["Cluster"] = labels

# 7. Evaluate clustering
silhouette = silhouette_score(X_scaled, labels)
dbi = davies_bouldin_score(X_scaled, labels)
print("\n=== CLUSTERING METRICS ===")
print("Silhouette Score :", round(silhouette, 3))
print("Davies–Bouldin   :", round(dbi, 3))

# 8. Cluster centers (original units)
centers = scaler.inverse_transform(km.cluster_centers_)
centers_df = pd.DataFrame(centers, columns=["Income_$", "SpendingScore"])
centers_df.index.name = "Cluster"
print("\n=== CLUSTER CENTERS (Original Units) ===")
print(centers_df.round(2))

# 9. Sanity check (3 sample rows)
print("\n=== SANITY CHECK SAMPLE ===")
print(df[["Income_$", "SpendingScore", "Cluster"]].sample(3))

# 10. Save labeled dataset
df.to_csv("spending_labeled_clusters.csv", index=False)
```
